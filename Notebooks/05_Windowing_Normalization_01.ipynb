#Windowing and Normalization

```python
import os
import numpy as np
```

##Train set windowing

```python
# paths
input_dir = "/content/brain_ct_project/train/images"
output_dir = "/content/brain_ct_project/train/images_windowed_normalized"

os.makedirs(output_dir, exist_ok=True)

# window params
WL = 40
WW = 80

lower = WL - (WW / 2)
upper = WL + (WW / 2)

print("Processing Train...")

count = 0

for file in sorted(os.listdir(input_dir)):

    if not file.endswith(".npy"):
        continue

    path = os.path.join(input_dir, file)

    try:
        img = np.load(path).astype(np.float32)
    except:
        print("Skipping:", file)
        continue

    # ---- windowing ----
    img = np.clip(img, lower, upper)

    # ---- normalize to 0–1 ----
    img = (img - lower) / (upper - lower)

    # ---- save ----
    np.save(os.path.join(output_dir, file), img)

    count += 1

print("Train done:", count)
```

#Remove the raw images

```python
import shutil

base = "/content/brain_ct_project"

for split in ["train", "val", "test"]:
    raw_path = os.path.join(base, split, "images")

    if os.path.exists(raw_path):
        shutil.rmtree(raw_path)
        print(f"Removed {raw_path}")
    else:
        print(f"Not found: {raw_path}")
```

```python
for split in ["train", "val", "test"]:
    base = f"/content/brain_ct_project/{split}"

    old = os.path.join(base, "images_windowed_normalized")
    new = os.path.join(base, "images")

    if os.path.exists(old):
        os.rename(old, new)
        print(f"Renamed in {split}")
```
