#Transform the task

##Train

```python
split = "train"
```

```python
import os
import numpy as np
import pandas as pd

base_path = "/content/brain_ct_project_data"

mask_dir = os.path.join(base_path, split, "masks")

data = []

for mask_name in sorted(os.listdir(mask_dir)):

    if not mask_name.endswith(".npy"):
        continue

    mask_path = os.path.join(mask_dir, mask_name)

    try:
        mask = np.load(mask_path)
    except:
        print(f"Skipping unreadable: {mask_name}")
        continue

    # binary label
    label = 1 if np.any(mask > 0) else 0

    data.append([mask_name, label])

df = pd.DataFrame(data, columns=["filename", "label"])

csv_path = os.path.join(base_path, split, "labels.csv")
df.to_csv(csv_path, index=False)

print("Saved:", csv_path)
print(df["label"].value_counts())
```
